## 1 Scrabble Points for past Wordles

```python
import pandas as pd
import censoring
```

```python
words_df = pd.read_csv(censoring.scrabble_path)
```

```python
# https://stackoverflow.com/questions/59722407/python-scrabble-word-count
score = {"a": 1, "c": 3, "b": 3, "e": 1, "d": 2, "g": 2,
         "f": 4, "i": 1, "h": 4, "k": 5, "j": 8, "m": 3,
         "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
         "r": 1, "u": 1, "t": 1, "w": 4, "v": 4, "y": 4,
         "x": 8, "z": 10}

def scrabble_word_count(word):
    res = 0
    for letter in word:
        res += score[letter]
    return res
```

```python
words_df['ScrabblePoints'] = words_df['Answer'].apply( lambda x: scrabble_word_count(x) )

words_df = words_df.astype({'PuzzleNum': 'int64'})
```

## 2 Player Wordle Scores
This project investigates reading the Wordle Scores shared to the 'Age of Among Us' _WhatsApp_ Chat

```python
import re
import pandas as pd
import censoring
```

```python
wordle_match = re.compile(r"(\w+ordle)")
template = 'date, time - name: wordle'
date_list, time_list, name_list, wordle_list = list(), list(), list(), list()
```

```python
with open(censoring.input_path) as input_file:
    for line in input_file:
        match = wordle_match.search(line)
        try:
            match_tuple = match.groups()
            # print(line)
            
            try: 
                s1 = line.split(', ')
                date_list.append(s1[0])
                
            except IndexError:
                s1 = template.split(', ')
                date_list.append(s1[0])
            
            try: 
                s2 = s1[1].split(' - ')
                time_list.append(s2[0])
                
            except IndexError:
                s2 = template.split(', ')[1].split(' - ')
                time_list.append(s2[0])
            
            try:
                s3 = s2[1].split(': ')
                name_list.append(s3[0])
                
            except IndexError:
                s3 = template.split(', ')[1].split(' - ')[1].split(': ')
                name_list.append(s3[0])
                
            wordle_list.append(s3[1])         
          
        except AttributeError:
            pass

# create dictionary
wordle_dict = {'Date': date_list, 
              'Time': time_list, 
              'Name': name_list, 
              'Wordle': wordle_list}
```

```python
# write results to data frame
df = pd.DataFrame(wordle_dict)
# filter out errors
df = df[df['Wordle'] != 'wordle']
# split wordle and count the number of words
df['WordCount'] = df['Wordle'].apply(lambda x: len(str(x).split(' ')))
# filter out word counts that do not equal 3
df = df[ df['WordCount'] == 3]
# split the left over words into columns
df[['Game', 'PuzzleNum', 'Score']] = df['Wordle'].str.split(expand=True)
# generate final score
df['FinalScore'] = df['Score'].apply(lambda x: x[0])
# merge date and time, convert to date time?
df['Date_Time'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
# sigh more filtering
df = df[ df['Game'] == 'Wordle']
# df = df[ df['FinalScore'] != 'X'] 
df = df[ df['FinalScore'] != 'g']
# censor players
replace_list = list(censoring.name_alias.items())
for i in range(8):
    df = df.replace(replace_list[i][0], replace_list[i][1])
# create final df
final_df = df[['Date_Time', 'Name', 'Game', 'PuzzleNum', 'FinalScore']]
# change types
final_df = final_df.astype({ 'Date_Time': 'datetime64', 'PuzzleNum': 'int64'})
```

## 3 Join Scrabble Point / Words to Player Scores

```python
final_df = final_df.merge(words_df[['PuzzleNum','Answer','ScrabblePoints']], on=['PuzzleNum'], how='left')
```
